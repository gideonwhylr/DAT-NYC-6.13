# Transforming Distributions

![](images/pandas/quet_book.jpeg)
A big reason for exploring the distributions of data is that our models assume a certain distribution (usually normal).  We will often be asking the question of whether certain variables are related to one another. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import california_housing
from pandas.plotting import scatter_matrix
```

```python
gap_all = pd.read_csv('data/gapminder_data/gapminder_all.csv')
```

```python
gap_all.info()
```

```python
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1)
plt.hist(gap_all.lifeExp_2007)
plt.title('Life Expectancy')

plt.subplot(1, 2, 2)
plt.hist(gap_all.gdpPercap_2007)
plt.title('GDP Per Capita')
```

```python
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1)
plt.hist(gap_all.gdpPercap_2007)
plt.title('GDP')

plt.subplot(1, 2, 2)
plt.hist(np.log(gap_all.gdpPercap_2007))
plt.title('GDP after Log Transform')
```

```python
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1)
plt.scatter(gap_all.gdpPercap_2007, gap_all.lifeExp_2007)
plt.title('Data Before Transformation')

plt.subplot(1, 2, 2)
plt.scatter(np.log(gap_all.gdpPercap_2007), gap_all.lifeExp_2007)
plt.title('Data After Log Transform')
```

Numerical values for these measures touch on how off center the distribution is (**skew**) and how peaked the distribution is (**kurtosis**).

```python
gap_all.gdpPercap_2007.skew()
```

```python
np.log(gap_all.gdpPercap_2007).skew()
```

```python
gap_all.gdpPercap_2007.kurtosis()
```
